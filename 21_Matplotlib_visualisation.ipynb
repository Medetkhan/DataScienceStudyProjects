### Шаг 3

Пожалуйста, создайте графики функций синуса, косинуса. На графике должно быть равномерно распределено 100 точек, в диапазоне (-10, 10).

````python
#Начало вашего кода
import math
np.random.seed(1)
points = np.linspace(-10,10, num=100)
s = np.sin(points)
c=np.cos(points)
plt.plot(points,s, label='sin')
plt.plot(points,c, label = 'cos')
plt.xlabel('xaxis')
plt.legend()
plt.title('Trigonometric functions')
plt.axhline(y=0, color='black')##add x axis on y=0
#Конец вашего кода
````
