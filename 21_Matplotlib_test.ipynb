```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
X=np.linspace(0,10,11)
```

```python
print(X)
```

```python
y=X
```

```python
plt.plot(X,y)
```

```python
plt.plot(X,X, '<--g', label='y=X', linewidth = 5, markersize=10);
plt.plot(X,X*X, '-.b', label='y=X^2', markersize = 9);
plt.plot(X,X*X*X, 'o:y', label = 'y=X^3', linewidth = 5, markersize = 10);
plt.legend();
plt.title('Plots of functions')
plt.xlabel('x-axis')
plt.ylabel('y-label')
```
